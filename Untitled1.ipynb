```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Visualization

Question1: How do filmmakers determine realease dates for different types of films?

Visualization: Count Plot of Genres by Release Month. \
A count plot shows the frequency of films of each genre released during each month.

```python
# Assuming df is your DataFrame containing the movie data

# Step 1: Extract release month from the 'Release Date'
df["Release Month"] = pd.to_datetime(df["Release Date"], errors="coerce").dt.month

# Step 2: Create a plot for the count of movies released by month and genre
plt.figure(figsize=(14, 8))

# Plotting the count of movies with seaborn countplot
sns.countplot(data=df, x="Release Month", hue="Major Genre", palette="Set3")

# Diagram title
plt.title("Anzahl der Filme nach Monat und Genre", fontsize=16)

# Axis labels
plt.xlabel("Verffentlichungsmonat", fontsize=14)
plt.ylabel("Anzahl der Filme", fontsize=14)

# Legend
plt.legend(title="Genre", bbox_to_anchor=(1.05, 1), loc='upper left')

# Customize x-ticks to show month names
plt.xticks(range(0, 12), 
           ["Jan", "Feb", "Mrz", "Apr", "Mai", "Jun", "Jul", "Aug", "Sep", "Okt", "Nov", "Dez"], rotation=45)

# Display the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Annahme: df ist Ihr DataFrame mit den relevanten Daten
# Verffentlichungsmonat extrahieren
df["Verffentlichungsmonat"] = pd.to_datetime(df["Release Date"], errors="coerce").dt.month

# Nach Monat gruppieren und den durchschnittlichen weltweiten Bruttoertrag berechnen, in Millionen umrechnen
durchschnittlicher_umsatz = df.groupby('Verffentlichungsmonat')['Worldwide Gross'].mean().reset_index()
durchschnittlicher_umsatz['Worldwide Gross'] = durchschnittlicher_umsatz['Worldwide Gross'] / 1_000_000  # In Millionen umrechnen

# Diagrammgre festlegen
plt.figure(figsize=(10, 6))

# Liniendiagramm erstellen
plt.plot(durchschnittlicher_umsatz['Verffentlichungsmonat'], 
         durchschnittlicher_umsatz['Worldwide Gross'], 
         marker='o', 
         label='Durchschnittlicher weltweiter Umsatz')

# Achsen beschriften
plt.xticks(ticks=range(1, 13), labels=['Jan', 'Feb', 'Mr', 'Apr', 'Mai', 'Jun', 'Jul', 'Aug', 'Sep', 'Okt', 'Nov', 'Dez'])
plt.xlabel('Verffentlichungsmonat')
plt.ylabel('Durchschnittlicher weltweiter Umsatz (in Millionen)')
plt.title('Durchschnittlicher weltweiter Umsatz nach Verffentlichungsmonat')

# Legende hinzufgen
plt.legend(fontsize=12)

# Raster anzeigen
plt.grid()

# Diagramm anzeigen
plt.show()
```

Question 2: How do ratings relate to box office success?\
Well plot IMDB ratings and Rotten Tomatoes scores against Box Office Revenues.

```python
import matplotlib.ticker as ticker

# Diagramm 1: Box-Office-Einnahmen nach IMDb-Bewertungen
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='IMDB Rating', y='Worldwide Gross', label='IMDb-Bewertung', color='blue')

# Y-Achsen-Beschriftung auf Millionenwerte anpassen
plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: f'{x*1e-6:.1f}M'))
plt.ylabel("Weltweite Einnahmen (in Millionen)")
plt.title("Box-Office-Einnahmen nach IMDb-Bewertungen")
plt.xlabel("Bewertung")
plt.legend(title="Bewertungsmetrik", loc="upper left")
plt.grid()
plt.show()

# Diagramm 2: Box-Office-Einnahmen nach Rotten-Tomatoes-Bewertungen
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='Rotten Tomatoes Rating', y='Worldwide Gross', label='Rotten-Tomatoes-Bewertung', color='green')

# Y-Achsen-Beschriftung auf Millionenwerte anpassen
plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: f'{x*1e-6:.1f}M'))
plt.ylabel("Weltweite Einnahmen (in Millionen)")
plt.title("Box-Office-Einnahmen nach Rotten-Tomatoes-Bewertungen")
plt.xlabel("Bewertung")
plt.legend(title="Bewertungsmetrik", loc="upper left")
plt.grid()
plt.show()
```

Question 3: Influence of Production Budget on Box Office Success.\
Well analyze the relationship between Production Budget and Box Office Revenue, segmented by genre.\

Average Box Office Revenue by Production Budget Categories. Visualize how average box office revenue changes with increasing production budget categories.

```python
import matplotlib.ticker as ticker
import seaborn as sns
import matplotlib.pyplot as plt

# Custom binning for production budget
df['Budget Kategorie'] = pd.cut(
    df['Production Budget'],
    bins=[0, 10e6, 50e6, 100e6, 300e6],  # Custom bin ranges
    labels=['Niedrig', 'Mittel', 'Hoch', 'Sehr Hoch']
)

# Calculate average worldwide gross revenue and count for each budget category
durchschnittlicher_umsatz = df.groupby('Budget Kategorie')['Worldwide Gross'].mean().reset_index()
filmanzahl = df['Budget Kategorie'].value_counts().reindex(durchschnittlicher_umsatz['Budget Kategorie']).values
durchschnittlicher_umsatz['Filmanzahl'] = filmanzahl

# Plotting the bar plot
plt.figure(figsize=(12, 6))
ax = sns.barplot(data=durchschnittlicher_umsatz, x='Budget Kategorie', y='Worldwide Gross', palette="viridis")

# Annotate bars with average revenue and movie count
for p, count in zip(ax.patches, durchschnittlicher_umsatz['Filmanzahl']):
    ax.annotate(f'{p.get_height() * 1e-6:.1f}M\n({count} Filme)',  # Revenue in millions + count
                (p.get_x() + p.get_width() / 2, p.get_height()),
                ha='center', va='bottom', fontsize=10)

# Format y-axis to show values in millions
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: f'{x*1e-6:.1f}M'))
# Add a legend for the budget categories
plt.legend(title='Budget Kategorie', labels=['Niedrig', 'Mittel', 'Hoch', 'Sehr Hoch'], loc='upper left')


# Set plot details
plt.title("Durchschnittlicher Box-Office-Umsatz nach Produktionsbudgetkategorie")
plt.xlabel("Produktionsbudgetkategorie")
plt.ylabel("Durchschnittlicher Box-Office-Umsatz (in Millionen)")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Add subtle gridlines for better readability
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

# Setze die Gre der Grafik
plt.figure(figsize=(10, 6))

# Streudiagramm zwischen Produktionsbudget und weltweitem Umsatz (Standardfarbe)
sns.scatterplot(data=df.dropna(subset=['Production Budget', 'Worldwide Gross']), 
                x='Production Budget', 
                y='Worldwide Gross', 
                label='Datenpunkte')  # Hinzufgen eines Labels fr die Legende

# Hinzufgen einer einfachen Regressionslinie
sns.regplot(data=df.dropna(subset=['Production Budget', 'Worldwide Gross']),
            x='Production Budget', 
            y='Worldwide Gross', 
            scatter=False,  # Entferne die Streudiagrammpunkte (werden bereits angezeigt)
            line_kws={'color': 'red', 'lw': 2},  # Rote Regressionslinie
            ci=None,  # Entferne den Konfidenzintervall-Schattierungsbereich
            label='Regressionslinie')  # Hinzufgen eines Labels fr die Legende

# Formatieren der Achsen, um Werte in Millionen darzustellen
plt.gca().xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: f'{x*1e-6:.1f}M'))
plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: f'{x*1e-6:.1f}M'))

# Titel und Achsenbezeichner festlegen
plt.title('Produktionsbudget vs. weltweiter Umsatz mit Regressionslinie')
plt.xlabel('Produktionsbudget (in Millionen)')
plt.ylabel('Weltweiter Umsatz (in Millionen)')

# Legende hinzufgen
plt.legend()

# Gitterlinien fr bessere Lesbarkeit aktivieren
plt.grid(True)

# Diagramm anzeigen
plt.show()
```
